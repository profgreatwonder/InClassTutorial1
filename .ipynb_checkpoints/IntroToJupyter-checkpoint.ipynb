## List Comprehension
this is a way of generating and specifying a list without having to iterate over all of them

```python
list1
```

```python
f = open("shakespare.txt") # this is to open a file called "shakespare.txt"
lines = f.readlines() # to read all the lines. lines will be the array of every line in this text file.
f.close() # it is important to close files as soon as you open them
```

```python
#the error above is because such file doesn't exist here.
```

```python
with open("shakespare.txt") as f:
    lines = f.readlines()  #what "with" does for you is that it closes the file as soon as it is done with the file.
```

```python
lines[0:5] #this prints out 5 lines of the file
```

```python
#to get the total number of characters
num_chars = 0
for line in lines:
    num_chars += len(line)
num_chars
```

```python
#I could repeat the code above using list comprehension

[len(line) for line in lines] #or we can add the sum function to add them all up sum([len(line) for line in lines])

#to get a list of words contained in each line -> [len(line.split(" ")) for line in lines] 

# to get the total number of words, you add the sum function sum([len(line.split(" ")) for line in lines])
```

```python
# to insert new words

%%timeit # this is known as cell magic. Will tell us how long every loop takes

# %%time this magic returns the total time taken

all_words = []
for line in lines[:10000]:
    for word in line.split():
        all_words.insert(0, word) #insert the words in reverse order always inserting new words at the beginning of list
        #to append, we have all_words.append(word)
```

## Classes

```python
# how to call a class
class myClass:
    # classes have a constructor, so you define an init function which a special function. This function takes 2 args.
    def _init_(self, n):
        # the n arg is like an initialization argument
        self.n = n # self._n = n to make it private
        # We need to find a class. For every function in the class, the first arg is always the object of the class
        # of the class itself. In self.n = n, self will always be the copy of itself. any other function within the
        # class will always be the class object itself
        
    # a function that gets what n is
    def get_n(self):
        return self.n # self._n to access private
```

```python
# to initialize

a = myClass(4)
a.n # or a.get_n() # a._n(to access private variables)

# every time the method of a class is called, that class will call the self argument as the first arg in all of the #
# in all of the functions

# In python, there's no such thing as private/public class. In python, everything is exposed. You can access all the #
# variables in any class we can always call a.n cos n is a memeber of the class, myClass object

# to define a hidden variable, you define them, self._n = n (instead of just self.n = n)

# we will be using classes as a way of encapsulating multiple fuunctions
```
